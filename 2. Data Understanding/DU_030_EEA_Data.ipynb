```python
df = pd.read_csv(os.path.join(DIR, files[2]), encoding = "utf-16")
df
```

```python
def map_to_datetime(row): 
    datetime_date = datetime.datetime.strptime(row[0:10], '%Y-%m-%d').date()
    return datetime_date 
```

```python
df["DateBegin"] = df.DatetimeBegin.apply(map_to_datetime)
```

```python
df["DateEnd"] =  df.DatetimeEnd.apply(map_to_datetime)
```

```python
df
```

```python
df["Day_of_year"] = df.DateBegin.apply(lambda x: x.timetuple().tm_yday)
df.Day_of_year.values
```

```python
for day in range(0, 366): 
    if day not in df.Day_of_year.values: 
        print(day)
```

```python
df["Weekday"] = df.DateBegin.apply(lambda x: x.weekday())
```
